```python
import os
import os.path as osp
import numpy as np
import pickle
import matplotlib.pyplot as plt
import numpy.linalg as LA

experiment_name = "drone_xz"
logdir = "/home/desong/Shaohang/plot_states"
```

```python
with open(osp.join(logdir, "run_stats.pkl"), "rb") as f:
    data = pickle.load(f)   
train_stats = data["train_stats"]

start = -1
num_traj = 200

for j in range(start, start-num_traj, -1):
    traj = train_stats[j]
    xs = []
    zs = []
    for i in range(len(traj)-1):
        step_stats = traj[i]
        state = step_stats["state"]
        xs.append(state[0])
        zs.append(state[1])
        # print(state)
    
    
    last_state = step_stats["state"]
    fig, axs = plt.subplots(1, figsize=(8, 8))
    axs.plot(xs, zs, '-')
    axs.plot(xs[0], zs[0], 'r*', markersize=12) # start point
    axs.plot(xs[-1], zs[-1], 'bo', markersize=12) # end point
    print("=============== trajectory {} ===============".format(j))
    print("end velocity: vx=", last_state[2], " vz=", last_state[3], "mag=", np.sqrt(last_state[2]**2 + last_state[3]**2))
    print("distance:", LA.norm(last_state[:2]-np.array([0, 0.5])))
    print("epi_steps: ", len(traj))
    rectangle_lower = plt.Rectangle((-1, 0.1), 0.5, 0.9)
    rectangle_higher = plt.Rectangle((0, 1.3), 1, 0.5)
    ground = plt.Rectangle((-4, 0), 5, 0.2)
    axs.set_ylim(0, 1.8)
    axs.set_xlim(-3, 1)
    plt.gca().add_patch(rectangle_lower)
    plt.gca().add_patch(rectangle_higher)
    plt.gca().add_patch(ground)
    plt.show()
```
